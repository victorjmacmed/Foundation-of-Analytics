# Introduction to tree models

Alternative way of modeling binary (or multinomial) problems. Instead of using a classification model (or logistic model) we use a binary tree (if the problem is binary - this is a decisition tree).

## Entropy impurity uncertainty

We call entropy the expression

$$H = -\sum_{i=1}^{c}p_i\log(p_i)$$

where $c$ is a non-negative integer.

### Example: informal example

1 - suppose we have an experiment $[1,1,1,\dots,1] where $ $c = 2$, $p_0 = 0$ and $p_1 = 1$, then $H = 0$ if we think about the first term as a limit 
term, that is, $\lim_{t \rightarrow 0}t\log(t) = 0$

2 - Another example: $[1,1,0,1,0,\dots,1]$, $p_0 = 1/2$ and $p_1 = 1/2$. We have $H = -2\frac{1}{2}\log_{2}(1/2) = 1$.

For the previous data sets 1 is pure and 2 is not pure. Entropy is a measure of impurity. 

# Classification tree algorithm


Suppose we have a table with 3 variables. two variables are categorical of type 1 or 0 (yest or no). Our table is 

| Play | Weather | windy |
| --- | --- | --- |
| 1 | sunny | no |
| 1 | sunny | yes |
|$\vdots$ | $\vdots$ | $\vdots$ | 

This table contains information about a particular person playing tennis or not certain day. We can write a decision tree based on this information. 

![alt text](trees.png "Title")

The questions here are about weather or windy days. 

__Example:__ John plays tennis 

| Play | Weather | windy |
| --- | --- | --- |
| 1 | sunny | no |
| 1 | sunny | yes |
|$\vdots$ | $\vdots$ | $\vdots$ | 

play variable: 20 points
weather - 15-1 and 5-0 (he plays, happens 15 times, and he does not play 5 times)
sunny - 12 sunny - 8 not sunny (on the sunny days he plays 8 days, on the not sunny days he plays 7 days and he does not play 5 days.) 

Sunny: 8-1, 0-0
not sunny: 7-1, 5-0

If we split by the windy conditions:

10 windy -> 6-1 , on windy 4-0 plays 6 times does not play 4 times
10 not windy - on not windy he plays 9-1, he does not play 1-0 


With this information we can calculate all the different probabilities. Notice that
$p_1 = 15/20$ and $p_2 = 5/20$.

\begin{align*}
H(parent) = -p_1\log(p_1)-p_0\log(p_0) = -\frac{15}{20}\log(\frac{15}{20})-\frac{5}{20}\log(\frac{5}{20})\approx 0.562.
\end{align*}

\begin{align*}
H(\text{sunny}) &= 0 \\
H(\text{not sunny}) &= 0.679.
\end{align*}

\begin{align*}
H(\text{childs, weather}) &= 0.407 \\
H(\text{windy,child}) &= 0.489.
\end{align*}

These are approximations. For weather condition you will get smaller entropy at the child node. The best question to ask is weather condition.


<img src="model.png" alt="drawing" width="500"/>

Some parameters we can control to avoid overfitting. The size of the data set should be bigger than 1000, the depth should be smaller than 10 or 5. We don't to many splits. 

__Summary of the steps:__

1. At each step the algorithm will check every variable for a possible split
2. A variable is selected for the split if it maximally reduces the impurity in the child nodes (e.g.
   the largest reduction of entropy, the largest reduction of variance etc.)
3. Pruning: a tree can grow till each leave only contains one data point. Pruning the tree is needed to avoid overfitting.

__Stopping criterion:__

1. The number of training sample after split is less than certain number ’min samples split’.
2. When the depth of the tree reaches to certain threshold ’max depth’.

# Building a single tree from scratch (using theoretical ideas)

Recall that 
$$H = -\sum_{i=1}^{c}p_i\log(p_i)$$

(we use - in the entropy, this comes from a logarithm and p < 1, therefore ... crystal clear)

```python
def entropy(y): # entropy function - binary variables
    p1 = float(y.sum())/float(len(y))+1e-20 # how many 1's divided by the total and a perturbation to avoid zeros (log)
    return -(p1*np.log2(p1)+(1-p1)*np.log2(1-p1)) # log_2 base 2
```

```python
entropy(ysimp) # entropy at the root/parent node
```

```python
def Dentropy(data, y, splitter): #entropy change after splits D from increase (Df 'increase/variation of f')
    yleft = y[data[splitter]<0.5] # data points that goes to the left
    yright = y[data[splitter]>=0.5] # data points that goes to the right
    H2 = (len(yleft)*entropy(yleft)+len(yright)*entropy(yright))/float(len(yleft)+len(yright)) # average entropy after splitting
    H1 = entropy(y) #entropy before splits
    dH = H1-H2 # increase - H2 child nodes
    return (dH, entropy(yleft), entropy(yright))

# Ideas here

# Remark: take into consideration that H is negative therefore reduction means being as positive as possible
#
# First we choose the root
# Then we test all the possible variables - we calculate the 'mean entropy' and dH 
# Then we take the variable with biggest entropy reduction as leaves

# This function calculate these changes/ This is a binary model
# This return a 'matrix'
```

# Random Forests

Binary models could overfit easily. The CART model is a single tree model but we can build a random forest.
In the car model, the binary tree is deterministic (we want to maximize the entropy reduction).

We can use particular parts of the dataset to build different trees. We randomly sample the dataset and build different trees. Then we get a forest.

<img src="forest.png" alt="drawing" width="200"/>


Once we have all these trees we want to find a way to combine them. The way to combine these trees is 'ask them to vote'. We have a value for a particular value 0 or 1, then we colect the ones or zeros from each tree.

The general idea is

1. We use a random sub-sample with replacement, to grow each individual tree.
2. At each node M << m columns are randomly selected, out of all m predictors. Usually 
   the default choice is $M = m/3$ for regression trees and $M = \sqrt(m)$ for classification trees.
3. Each tree is grown to the largest extent possible (one sample in the leaves)

# Implementing random forest to predict house pricing.

```python
from sklearn.ensemble import RandomForestRegressor
import sklearn
```

```python
houseprice = pd.read_csv('housetrain.csv')
```

```python
houseprice
```

```python
houseprice.shape
```

```python
houseprice.columns
```

```python
y = houseprice['SalePrice']
predictor_cols = ['LotArea','OverallQual','OverallCond','YearBuilt']

# Create training set - predictors

train_X = houseprice[predictor_cols]
```

```python
# sampling

rf = RandomForestRegressor(n_estimators = 50, max_features = 'sqrt', max_samples = 0.8, max_depth = None, random_state = 42)
# Parameters

# n_estimator -> number of trees
# max_features -> how many features we want to consider - sqrt(m) m = total number of columns
# max_samples -> 400 data points we take 80% to grow a single tree
# max_depth -> None max possible tree (without limitations) -> depth of the tree
# random_state -> We want the samples to be randomly selected (we get same results all the time even if this is random)

# n_estimator is an importan parameter. We want to take this parameter accurately in order to reduce error.
# For instance if we take this parameter to be 1000 the error will be smaller but we could be overfitting.
# We have to use crossvalidation to reduce overfitting. We want to choose the parameters such that the error
# is minimum and we are not overfitting.

rf.fit(train_X,y);
```

```python
rf.estimators_ # 50 trees here 
```

```python
tree.plot_tree(rf.estimators_[1])
```

```python
# Here we have 50 different trees
# We are doing a regression.
# We will apply all these trees to our predictors
```

```python
rf.feature_importances_
```

```python
houseprice['ypred'] = rf.predict(train_X)
```

```python
error = np.sqrt(np.dot((y-houseprice['ypred']), (y-houseprice['ypred'])))
error
```

```python
houseprice.plot(x = 'OverallQual', y = ['SalePrice', 'ypred'], figsize = (10,5), grid = True, style = ['o','rx'])
plt.title('SalePrice scatter')
plt.show()

# Our predictions are good enough
```

# Improving the previous model (optimal number of parameters)

## Avoiding overfitting

```python
# n_estimator is an importan parameter. We want to take this parameter accurately in order to reduce error.
# For instance if we take this parameter to be 1000 the error will be smaller but we could be overfitting.
# We have to use crossvalidation to reduce overfitting. We want to choose the parameters such that the error
# is minimum and we are not overfitting.

from sklearn.model_selection import KFold
from sklearn.utils import shuffle

# create k-folds and shuffle

kf = KFold(n_splits = 10)
houseprice_shuffle = shuffle(houseprice)

# taking the 'shuffled' y variable 

y_shuffle = houseprice_shuffle['SalePrice']
predictor_cols = ['LotArea','OverallQual','OverallCond','YearBuilt']

# Create training predictors data

train_X_shuffle = houseprice_shuffle[predictor_cols]
```

```python
errorls = []

for ntrees in [10, 50, 100, 500, 1000]:
    y_pred_cv = np.empty((0,1))
    rf = RandomForestRegressor(n_estimators = ntrees, max_depth=None, min_samples_leaf=1)
    for train, test in kf.split(train_X_shuffle):
        rf_modelcv = rf.fit(train_X_shuffle.iloc[train], y_shuffle.iloc[train]);
        y_pred_cv = np.append(y_pred_cv, rf_modelcv.predict(train_X_shuffle.iloc[test]))
    errorls.append(np.sqrt(np.dot(y_pred_cv-y_shuffle, (y_pred_cv-y_shuffle)/len(y_shuffle))))
    
    
# Train with a sample k-fold sample.
# Then calculate predictions and add these predictions to the same array.
```

```python
houseprice_shuffle['y_pred_cv'] = y_pred_cv
houseprice_shuffle['residual'] = houseprice_shuffle['y_pred_cv']-houseprice_shuffle['SalePrice']
```

```python
houseprice_shuffle.plot(x = 'OverallQual', y=['residual'], figsize = (10,5), grid = True, style = ['o','rx'])
plt.title('SalePrice scatter')
plt.show()
```

```python
plt.plot(errorls)
plt.show()

# The error decrease when we increase the number of trees until 500 to 1000 then the error increase.
# Using this code we can decide the number of trees in our model 
```
