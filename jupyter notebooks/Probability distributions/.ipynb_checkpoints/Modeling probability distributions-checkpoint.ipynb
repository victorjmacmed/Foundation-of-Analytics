# Constructing Binomial from Bernoulli

Recall that the Binomial variables are just the sum of independent Bernoulli variables. Here we construct a Binomial variable from Bernoulli random variables.

```python
np.random.binomial(1,0.5,3) # simulate Bernoulli 3 times (p = 0.5)
```

```python
bernoulliconfig = np.empty((5000,3)) # this is a matrix 5000x3

for i in range(5000):
    bernoulliconfig[i,:] = np.random.binomial(1,0.5,3)

# each row, one simulation of a Bernoulli variable
```

```python
binomial_simulated = np.sum(bernoulliconfig, axis= 1) # sum by row (axis = 1)
```

```python
uniquevale_3, elements_count_3 = np.unique(binomial_simulated, return_counts= True)
```

```python
uniquevale_3
```

```python
elements_count_3/5000
```

```python
binblock_2 = np.arange(-0.5,4,1) # from -0.5 to 4 with increments of 1
plt.hist(binomial_simulated, bins = binblock_2, rwidth = 0.9)
plt.title('histogram')
plt.show()
```

```python
np.mean(binomial_simulated)
```

```python
np.var(binomial_simulated)
```

# Vectorize

Idea: instead of writing sum using a for we can use matrix multiplication.

```python
M = np.random.rand(5000,5000) # uniform distribution [0,1]
```

```python
M.shape
```

```python
np.sum(M[0,:])
```

```python
np.ones(5000)
```

```python
np.dot(M[0,:], np.ones(5000))
```

```python
np.matmul(M,np.ones((5000,1)))
```

```python
for i in range(5000):
    print(np.sum(M[i,:]))
```

```python
# measuring the execution time

import time 
t0 = time.time()
np.matmul(M,np.ones((5000,1)))
t1 = time.time()
print(t1-t0)
```

```python

import time 
t0 = time.time()
for i in range(5000):
    (np.sum(M[i,:]))
t1 = time.time()
print(t1-t0)

# This code is slower. Python matrix multiplication is optimized.
# Conclusion: we should always vectorize our operations.
# Another remark: GPUs are optimized for vector operations.
```
